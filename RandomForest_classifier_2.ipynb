```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('kyphosis.csv')
```

```python
data.head()
```

```python
Kyphosis=pd.get_dummies(data["Kyphosis"],drop_first=True).astype(int)
```

```python
data.drop("Kyphosis",axis=1,inplace=True)
```

```python
data=pd.concat([data,Kyphosis],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop("present",axis=1)
y=data["present"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
classi=RandomForestClassifier(n_estimators=100)
classi.fit(X_train,y_train)
```

```python
y_pred=classi.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

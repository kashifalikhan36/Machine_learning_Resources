```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Employee_Salary.csv')
```

```python
data.head()
```

```python
sns.jointplot(y='Salary',x='Years of Experience',data=data)
```

```python
sns.lmplot(y='Salary',x='Years of Experience',data=data)
```

```python
sns.pairplot(data)
```

```python
y1=data[['Salary']]
x1=data[['Years of Experience']]
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(x1,y1,random_state=42,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True)
regressor.fit(X_train,y_train)
```

```python
y_predict=regressor.predict(X_test)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_predict)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error,mean_absolute_error,r2_score
MSE=mean_squared_error(y_test,y_predict)
MAE=mean_absolute_error(y_test,y_predict)
MAPE=mean_absolute_percentage_error(y_test,y_predict)
R2=r2_score(y_test,y_predict)
print(MSE)
print(MAE)
print(MAPE)
print(R2)
```

```python
data6=pd.read_csv('kc_house_data.csv')
```

```python
data6.head()
```

```python
data6.describe()
```

```python
sns.pairplot(data6)
```

```python
data6.drop('id',axis=1,inplace=True)
data6.drop('date',axis=1,inplace=True)
```

```python
f,ax=plt.subplots(figsize=(20,20))
sns.heatmap(data6.corr(),annot=True)
```

```python
x = data6[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'sqft_living15', 'sqft_lot15']]
```

```python
y=data6['price']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scalled=scaler.fit_transform(x)
```

```python
y_scalled=scaler.fit_transform(y.values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scalled,y_scalled,test_size=0.25)
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(50,input_dim=14,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
epochs_hist=model.fit(X_train,y_train,epochs=100,batch_size=50,validation_split=0.2)
```

```python
y_predict=model.predict(X_test)
plt.plot(y_test,y_predict,'^',color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
y_predict=scaler.inverse_transform(y_predict)
y_test=scaler.inverse_transform(y_test)
plt.plot(y_test,y_predict,'^',color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,mean_absolute_percentage_error
MSE=mean_squared_error(y_test,y_predict)
MAE=mean_absolute_error(y_test,y_predict)
MAPE=mean_absolute_percentage_error(y_test,y_predict)
R2=r2_score(y_test,y_predict)
print(MSE)
print(MAE)
print(MAPE)
print(R2)
```

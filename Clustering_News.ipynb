```python
  from sklearn.datasets import fetch_20newsgroups
  from sklearn.feature_extraction.text import CountVectorizer
  from sklearn.feature_extraction.text import TfidfTransformer
  from sklearn.naive_bayes import MultinomialNB
```

```python
category=['alt.atheism','soc.religion.christian','comp.graphics','sci.med']
train=fetch_20newsgroups(subset='train',categories=category,shuffle=True)
test=fetch_20newsgroups(subset='test',categories=category,shuffle=True)
```

```python
count_vector=CountVectorizer()
x_train_tf=count_vector.fit_transform(train.data)
print(x_train_tf)
```

```python
#TfidfVectorizer = CountVectorizer + TfidfTransformer
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer=TfidfTransformer()
x_train_tfidf=tfidf_transformer.fit_transform(x_train_tf)
print(x_train_tfidf)
```

```python
model= MultinomialNB().fit(x_train_tfidf,train.target)
```

```python
x_new_counts=count_vector.transform(test.data)
x_new_tfidf=tfidf_transformer.transform(x_new_counts)
```

```python
pred=model.predict(x_new_tfidf)
```

```python
new=['My favourite topic has something to do with quantum physics and quantum mechanics',
     'This has nothing to do with church or religion',
     'Software engineering is getting hotter and hotter nowadays']
```

```python
x_new_counts=count_vector.transform(new)
x_new_tfidf=tfidf_transformer.transform(x_new_counts)
pred_y=model.predict(x_new_tfidf)
```

```python
for doc,category in zip(x_new_tfidf,pred):
    print('%r ------------> %s'%(doc,train.target_names[category]))
```

```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def model(X_train,y_train,lr,iterations):
  m=X_train.shape[1]
  n=X_train.shape[0]

  w=np.zeros((n,1))
  b=0

  for i in range(iterations):
    hy=np.dot(w.T,X_train)+b   # predicted value
    pred=sigmoid(hy) #pi

    dw=(1/n)*np.dot(X_train,(hy-y_train).T)
    db=(1/n)*np.sum(hy-y_train)

    w=w-lr*dw
    b=b-lr*db

  return w,b

def predict(x, w, b):
    # Linear combination of inputs and weights
    linear_pred = np.dot(w.T,x)+b  # x has shape (499, 2), w should have shape (2, 1)

    # Apply sigmoid activation
    y_pred = sigmoid(linear_pred)
    y_pred = y_pred.flatten()
    class_pred = [0 if y <= 0.5 else 1 for y in y_pred]

    return class_pred
```

```python
data = pd.read_csv('Facebook_Ads_2.csv', encoding='ISO-8859-1')
x=data.drop(['Clicked','Names','emails','Country'],axis=1).values
y=data[['Clicked']].values
data.head()
```

```python
iteration=10000
lr=0.01
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
w,b=model(X_test,y_train,lr,iteration)
```

```python
y_pred=predict(X_test,y_test,b)
```

```python
def accuracy(y_pred,y_test):
  return np.sum(y_pred==y_test)/len(y_test)

acc=accuracy(y_pred,y_test)
print(acc)
```
